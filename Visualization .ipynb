```python
import pandas as pd
import matplotlib.pyplot as plt

iris = pd.read_csv('iris.csv' )
iris.columns = ['No.', 'sepal_length', 'sepal_width' , 'petal_length', 'petal_width', 'species']

```

## Scatter Plot


```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width') ;
plt.show()
```

###  Create color dictionary


```python
import seaborn as sns

sns.set_style("darkgrid");
sns.FacetGrid(iris, hue="species", height=8) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend();
plt.show();
```

Scatter plot with color-coding giving the information like "Setosa"  is easy to identify 

### Pair Plot

```python
# pairwise scatter plot: Pair-Plot
# Dis-advantages: Cannot visualize higher dimensional patterns in 3-D and 4-D. 
#Only possible to view 2D patterns.
plt.close();
sns.set_style("darkgrid");
sns.pairplot(iris, hue="species", size=3);
plt.show()
# NOTE: the diagnol elements are PDFs for each feature. PDFs are expalined below.
```

-----petal_length and petal_width are the most useful features to identify various flower types.

-----While Setosa can be easily identified (linearly seperable), Virnica and Versicolor have some overlap (almost linearly seperable).

## Histogram 

```python
iris.hist()
```

## Box Plot

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
```

## Violin Plot

```python
sns.violinplot(x='species', y='petal_length', data=iris)
```

## Kernel Density Estimate of the underlying feature (KDE Plot)

```python
sns.FacetGrid(iris, hue="species", size=8) \
   .map(sns.kdeplot, "petal_length") \
   .add_legend()
```
